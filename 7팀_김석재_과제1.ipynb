```python
class solve:
    def __init__(self,S,Q):
        #주어진 문자열 S,Q가 하나의 문자열일경우 for문 사용시 문자 하나하나를 자르기때문에 리스트화한다
        if type(S)==str:
            self.S=[S]
        else:self.S=S
        if type(Q)==str:
            self.Q=[Q]
        else:self.Q=Q

    #첫번째 문제
    def solve1(self):
    # Sentece가 Q안에 존재하면 yes 아니면 No
        answer=['No','Yes']
        for Q_sent in self.Q:                
            for Sentence in self.S:
                print(f'#Q = {Q_sent} : {answer[Sentence in Q_sent]}') #boolean의 true(1) 와 false(0)을 사용 리스트의 문자열을 출력

    #두번째 문제
    def solve2(self):
        for Q_sent in self.Q:
            for Sentence in self.S:
                answer=Q_sent.find(Sentence)# Q에서 Sentence를 찾고 시작위치를 반환한다
                if answer==-1:continue #찾지못하면 -1을 출력하게된다
                print(f'#Q = {Q_sent}, #pos = {answer}, pattern = {Sentence}')

    #세번째 문제
    def solve3(self):
        def check(q=0):
            answer=Q_sent[q:].find(Sentence) #이미 값을 찾은 시작점까지 없애준다        
            if answer>=0:            
                print(f'#Q = {Q_sent}, #pos = {answer+q}, pattern = {Sentence}')
                check(q+1+answer)#Q에서 시작위치값(answer)을 더해준다 +1을 안하면 0+0이되어서 무한루프
                
            else:pass

        for Q_sent in self.Q:
            for Sentence in self.S:
                check()  

```

```python
ans = solve(S,Q)
```

```python
ans.solve1()
```

```python
ans.solve3()
```

#테스트


```python
import random
import string
```

```python
#랜덤으로 문자열을 생성
def make_num(num,num2):
    test_list=[]
    text = ""
    
    while num>1: #문자열의 갯수를 다 소진할때까지    
        for i in range(random.randrange(1, num2)): #문자열의 길이
            text += str(random.choice(string.ascii_lowercase)) #소문자로만 생성
        test_list.append(text)        
        text = ""
        num-=1
    return test_list     
    
```

```python
s=random.randrange(1, 10)#문자열 갯수 실제로는 S는 최대1000,Q는 10000인데 눈으로 확인을 할수가없어서 낮춰서 테스트
q=random.randrange(1, 100)
S_test=make_num(s,10) #두번째 인자는 문자열 길이의 최대값으로 S는 100 Q는 10000
Q_test=make_num(q,100)
```

```python
ans_test = solve(S_test,Q_test)
```

```python
ans_test.solve1()
```

```python
ans_test.solve2()
```

```python
ans_test.solve3()
```
